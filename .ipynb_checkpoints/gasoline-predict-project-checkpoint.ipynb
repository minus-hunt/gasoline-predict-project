# Прогнозирование типа и расхода бензина

## Предварительная обработка и исследовательский анализ данных

```python
import pandas as pd # импорт иблиотеки Pandas
df = pd.read_csv("measurements.csv") # загрузка данных
```

```python
df.head()
```

```python
df.info()
```

### Импорт библиотек для визуализации и вывод первого графика

```python
# при визуализации возникает много предупреждений
import warnings # при визуализации возникает много предупреждений
# данная библиотека позволяет игнорировать их
warnings.filterwarnings("ignore") 
# библиотека визуализации
import seaborn as sns
# импорт модуля для корректирования размеров 
import matplotlib.pyplot as plt 
%matplotlib inline

# вывод посчитанных значений в столбце specials
plt.figure(figsize = (10, 4))
plt.title("Подсчитанные значения в столбце \"Specials\"")
sns.countplot(df["specials"])
```

```python
df["refill liters"].value_counts()
```

```python
df["refill gas"].value_counts()
```

#### Удаление ненужных столбцов тк они содержат менее 50% от общего количества строк в DataFrame

```python
df = df.drop(columns=["specials", "refill liters", "refill gas"], axis=1)
```

```python
df.head()
```

```python
df.info()
```

#### Замена запятых на точки

```python
df["distance"] = df["distance"].str.replace(",", ".")
df["consume"] = df["consume"].str.replace(",", ".")
df["temp_inside"] = df["temp_inside"].str.replace(",", ".")
```

#### Приведение к типу `float64`

```python
df["distance"] = df["distance"].astype("float64")
df["consume"] = df["consume"].astype("float64")
df["temp_inside"] = df["temp_inside"].astype("float64")
df["temp_outside"] = df["temp_outside"].astype("float64")
```

```python
df.head()
```

```python
df.info()
```

#### Заполнение значений `NaN` по медиане

```python
df["temp_inside"] = df["temp_inside"].fillna(df["temp_inside"].median())
```

#### Визуализация типов бензина

```python
plt.figure(figsize = (10, 6))
plt.title("Соотношение типов бензина которыми заправлялся автомобиль")
sns.countplot(df["gas_type"])
```

```python
df["gas_type"].value_counts()
```

### Замена значений на уникальный номер

```python
# Импортируется нужный модуль
from sklearn.preprocessing import LabelEncoder
# Создается экземпляр класса
le = LabelEncoder()
# Подмена занчений на уникальный код
df["gas_type"] = le.fit_transform(df["gas_type"])
```

```python
df["gas_type"].value_counts()
```

```python
df.info()
```

### Корреляционная карта

```python
plt.figure(figsize = (9, 9))
sns.heatmap(df.corr(), annot = True)
```

### Матрица рассеивания

```python
sns.pairplot(df, hue = "gas_type")
```

## Построение моделей машинного обучения с учителем

### Разделение на тестовую  и обучающую выборки

```python
# Импорт модуля который создает выборки 
from sklearn.model_selection import train_test_split

# Определение целевой переменной и признаков для классификации
x_classf = df.drop("gas_type", axis = 1)
y_classf = df["gas_type"].values

# Разделение для классификации
x_train, x_test, y_train, y_test = train_test_split(x_classf, y_classf, test_size = 0.1, shuffle = True)
```

### Импорт библиотек и модулей для обучения,  и обучение первой модели XGBClassifer

```python
from xgboost import XGBClassifier, plot_importance
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
# Импортируем метрики
from sklearn.metrics import classification_report

# Обучение первой модели XGBClassifer
xgb = XGBClassifier() # создается экземпляр
xgb.fit(x_train, y_train)
```

```python
xgb.score(x_test, y_test)
```

```python
plot_importance(xgb)
```

### KNeighborsClassifer

```python
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
```

```python
kn.score(x_test, y_test)
```

### DecisionTreeClassifer

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

### Оптимизация гиперпараметров

```python
# Создается список в который заносятся значения 
# передающеися в экземпляр класса
estims = [100, 200, 300]
# Создаются еще два пустых списка.....
# первый для оценки точности по тренировочной выборке
train_score = []
# второй для оценки точности по тренировочной выборке
test_score = []
for es in estims:
    # экземпляр с передаваемыми параметрами
    xgb = XGBClassifier(n_estimators=es, verbosity=1)
    xgb.fit(x_train, y_train)
    train_score.append(xgb.score(x_train, y_train))
    test_score.append(xgb.score(x_test, y_test))

# Вывод результатов
print("TRAIN", train_score)
print("TEST", test_score)
```
